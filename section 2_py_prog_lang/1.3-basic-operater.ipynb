Arithemic Operater

```python
# Arithemic operater
a=20
b=10 
add_results=a+b
sub_results=a-b
multi_results=a*b
division_results=a/b

floor_divison_results=a//b
moduls_results=a%b
expon_results=a**b

print(add_results)
print(sub_results)
print(multi_results)
print(division_results)
print(floor_divison_results)
print(moduls_results)
print(expon_results)
```

Logical operater and or not


```python
# and , both are necessay to true then true otherwise false
x=False
y=True
results1=x and y
print(results1)
```

```python
# or , if one conditon is true then the true
x=False
y=True
results1=x or y
print(results1)
```

```python
# not , the reverse true >=not  || false>=true
X=True
not X
print(X)
Y=False
not Y
print(Y)
```
