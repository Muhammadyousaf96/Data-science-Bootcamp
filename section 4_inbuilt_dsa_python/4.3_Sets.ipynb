Sets are also a inbuilt data structure which are store unique value, unordered and do not duplicate

```python
# creating of sets

sets=set()
print(type(sets))
sets.add(2)
print(sets)

set2={} #if crating sets by the bracket and the bracket is empty the this is also a dictonary not set
set3={'yousaf',4,True}
print(type(set2))
print(set3)
```

```python
# set into list and add element to list and then into sets
setl={1,2,3,4,5}
print(type(setl))

intolist=list(setl)
print(type(intolist))
# add item by using the list method 
intolist.append(3)
intolist.append('yousaf')
intolist.insert(3,'khan')

print(intolist)
```

```python
# Empty set 
Empty_set=set()
print(type(Empty_set))
```

```python
# operation like adding 
sets={8,2,3,4,5}
sets.add('tuple')
sets.add('Muhammad ')
sets.add("Youasf")
print(sets)

# removing 
sets.remove(5)
sets.remove("tuple")
print(sets)

# pop
sets.pop()
sets.pop()
```

```python
# clera
sets.clear()
print(sets)
```

```python
# set membership test: wether the an element are present are not 
setM={2,4,5,6,7,8,9,'a','b','c',True,1.5,2.5}
print(setM)
print(3 in setM)
print(1 in setM)
print('a' in setM)
print(True in setM)
```

```python
# Mathematical operatoin 

set1={1,2,3,4,5,6}
set2={6,7,8,9,10}
set3={1,2,3}
set4={6,7,8,3,2,1,0}

# Unoin
# print(set1.union(set2))


# intersection
print(set1.intersection(set3))
print(set1.intersection_update(set3))
print(f'set1 : {set1}')
print(f"set3 : {set3}")

# difference 

print(set2.difference(set3))

# symmetric difference 
print(set2.symmetric_difference(set1))
print(set3.symmetric_difference(set4))

```

```python
# set method 
set1={1,2,3}
set2={6,7,8,9,10}
set3={1,2,3,4,5}
set4={6,7,8,3,2,1,0}

# issubset
print(set1.issubset(set3))
print(set1.issubset(set4))

# issuperset
print(set3.issuperset(set1))
print(set4.issuperset(set1))
```

```python
# countig unique word in text 
text="This is the paragraph where we see the unique word in text. and is a good practice  unique word in text. and is a good practice to splice the paragraph "
word=text.split()
print(word)
# changing to unique word convert the text into set because set only give unoqu word
unique_word=set(word)
print(len(word))
print(f"Unique Word: {unique_word}")
print(len(unique_word))
```
