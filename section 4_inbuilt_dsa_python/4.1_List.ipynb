```python
# List Comprehension very important 
# list comprehension means to create a new list from exesition list in a one line 

# 3 type, simple,conditional,neated list comprehension
# simple or basic :: [exeperaton for iteam iterating]\

# [x**2 for x in range(20)]

[x for x in range(20)]

# with conditional logic same but the condition at last
[x**2 for x in range(19) if x%2==0]
```
