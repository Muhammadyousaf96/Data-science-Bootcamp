Tuple are also a builtin data structure in python which are simlier to list but the change are in their mutabality. means once created than can be change

```python
# crating tuple
tu=()
print(type(tu))
print(tu)
```

```python
# using string 
stuple=('name','khan')
print(stuple)
```

```python
# using list 
lst=[1,2,3,4,5]
list_tuple=tuple(lst)
print(type(lst))
print(f"the list into tuple : {list_tuple}")
print(type(list_tuple))
```

```python
# using built-in function
tpl=tuple()
print(tpl)
stpl=tuple()
```

```python
# creating tuple with mixed data type
mixed_tuple=(1,'yousaf',True,2.4)
print(mixed_tuple)
```

```python
# nested tuple
tup1=(1,2,3)
tup2=(5,7,8)
tup3=(tup1,tup2)
print(type(tup3))
```

```python
# tuple with repetation
tupleRep=(3,'you',6)
tuprep=tupleRep *3
print(tuprep)
```

```python
# creating tuple with use of for loop
tup=('Yousaf')
n=6
for i in range(int(n)):
    tup=(tup,)
    print(tup)
```
