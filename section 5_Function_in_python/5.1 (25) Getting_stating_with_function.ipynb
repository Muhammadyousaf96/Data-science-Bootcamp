Why function need

```python
# even odd number ---simple 
num=21
if num%2==0:
    print('even')
else:
    print('odd')    
```

```python
# even odd number ---with function
def even_odd_funcion(num):
    if num%2==0:
        print(f"The number {num} is Even")
    else:
        print(f"the Number {num} is odd")    
```

```python
even_odd_funcion(41)
even_odd_funcion(20)
```

```python
# function with multiple parameter
def multipy(a,b):
    c=a*b
    print(c)
multipy(2,6)    
```

positional argument 

```python
# positional argument 

# def print_numbers(num1,num2,....):
def print_numbers(*args):
    for number in args:
        print(number)
print_numbers(1,2,3,4,5,6,'print_numbers',[1,2,3],{1,2,3})
```

Keyword Arguments

```python
# keyword Arguments ** star
# all the parameter are in a key value pair

def details(**keargs):
    for key,val in keargs.items():
        print(f"{key}:{val}")
```

```python

details(name="muhammad Yousaf",age=21,gender="male")
```
