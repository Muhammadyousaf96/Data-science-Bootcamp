lambda funtion is a naoanmous funtion (funtion wo=ithout name )

```python
# syntax of lambda function
# lambda argument:expression

# adding two number
# simpe function
def addition(a,b):
    return a+b
addition(1,2)    
```

```python
# with lambda funtion
addition=lambda a,b:a+b
print(addition(3,4))
```

```python
add=lambda x,y,z:x+y+z
print(add(1,4,5))
```

```python
# even simple function
def even_number(num):
    if num%2==0:
        return True
even_number(30)    
```

```python
# even with lambda 
even_number=lambda num:num%2==0
even_number(6)
```

Lambda with map function

```python
#lambda + map()
numbers=[1,2,3,4,5,8,10]
# to square all the list element
def square(num):
    return num**2
square(2)
```

```python
# lambda +map
# map(function,num)
list(map(lambda x:x**2,numbers))
```
