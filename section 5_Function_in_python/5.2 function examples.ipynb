Example 2 : passsword strenght checker 

```python
def strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isuper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    else:
        return True

  
```

```python
print(strong_password('Weak_pass'))
print(strong_password('W1$bookh'))
print(strong_password('jghwgfyuewrfewh5646&'))
```

Example 3: calculate the total cost of item in a shapping cart

```python
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']*item['quantity']
    return total_cost    

cart=[
    {'name':'Apple','price':0.5,'quantity':4},
    {'name':'Banana','price':0.6,'quantity':8},
    {'name':'Mango','price':0.7,'quantity':12},
    {'name':'Orange','price':0.9,'quantity':3},
]    
# calling function 
total_cost=calculate_total_cost(cart)
print(total_cost)
```

Example 4 : Check if a string is pailadroem


```python
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]
print(is_palindrome('Moom'))
print(is_palindrome('hejej'))
is_palindrome('ssssss')

print(is_palindrome('SystemErro87b r'))
```

Example 5: calculate the factorial of recursive function

```python
def factorial(num):
    if num==0:
        return 1
    else:
        return num* factorial(num-1)
factorial(2)
```

Exapmle 6: functin read file and count frequence of each word

```python
def count_word_frequence(file_path):
    word_count={}
    with open(file_path,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('.,/;')
                word_count[word]=word_count.get(word,0)+1
            return word_count
filepath="sample.txt"
word_frequence=count_word_frequence(filepath)
print(word_frequence)        
```
