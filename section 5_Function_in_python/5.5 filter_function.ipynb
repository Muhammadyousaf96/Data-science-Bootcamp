Filter function: is used to select specific items from a list based on a condtiton 
filter(function,iterable)

```python
# even number filter
numbers=[1,2,4,6,8,99,45,3,21,63]

def even_number(number):
    return number%2==0

def not_even(number):
    return number%2!=0

even=list(filter(even_number,numbers))
not_even=list(filter(not_even,numbers))
print(even)
print(not_even)
```

```python
# filter with lambda 
# greater than 5 
Number=[1,4,7,9,34,56,2,3,4,0]
list(filter(lambda x:x>5,Number))
```

```python
# even and grater than five
Number=[1,4,7,9,34,56,2,3,4,0]

list(filter(lambda x:x>5 and x%2==0,Number))
```
