```python
# sqare a number 
numbers=[1,2,3,4,56,7]

def square(number):
    return number**2

list(map(square,numbers))
```

```python
# lambda + map
numbers=[1,2,3,4,56,7]

list(map(lambda x:x**2,numbers))
```

```python
# can map multiple iterable [1,2,3]+[4,5,6]=[5,7,9] the yes

number1=[1,2,3]
number2=[4,5,6]

list(map(lambda x,y:x+y,number1,number2))
```

```python
# map to convert list of string to intrger 
lst_string=['1','2','3','4','5']

list(map(int,lst_string))
```

```python
# inbuild data structure 
# str.upper to upper the itreeam 

str_upper=['apples','mangoo','oranges']
list(map(str.upper,str_upper))
```

```python
# apply to list of dictionary 

def get_name(person):
    return person['name'],person['age']
    
people=[
    {'name':'m yousaf','age':22},
    {'name':"M Danish",'age':3}
]

list(map(get_name,people))
```
