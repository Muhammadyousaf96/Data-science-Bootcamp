Here we practice loop in python like, for and while and break,continue and pass and practice some problem 

```python
# for loop used when you know how many time to iterate the loop 
range(3)
```

```python
for i in range(1,10,2):
    print(i)
    
```

```python
for i in range(10,4,-1):
    print(i)
```

```python
# use in string
str="muhammad yousaf"
for i in str:
    print(i)
```

```python
name="muhammad Yousaf"
for i in name:
    if i=='d':
        print("the vale is d :",i)
```

```python
# now use of break and continue And pass in for loop
# break , stop the loop  imeturally

for i in range(1,9):
    if i==5:
        break
    print(i)

```

```python
# continue skip the 
for i in range(1,9):
    if i==5:
        continue
    print(i)

```

```python
# pass placeholder 
for i in range(1,9):
    if i==5:
        print("the pass value :",i)
        pass
    print(i)

```

```python
# nested for loop
for i in range(1,7):
    for j in range(1,3):
        print(f" i:{i} and j:{j}")
```

Now while loop , use when dont khow how many time to iterate the conditon, depent on condition

```python
count=1
while count<5:
    print(count)
    count=count+1
```

```python
# print sum of first 10 natural digit 
sum=0
count=1

while count<11:
    sum=sum+count
    count=count+1
print(sum)
```
