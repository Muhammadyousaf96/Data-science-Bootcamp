python provide a built in function to work with file


Reading a file
===


```python
# reading a file 
with open("example.txt","r") as file:
    content=file.read()
    print(content)
  
```

Writting a file
===

```python
# witing a file 
with open('example.txt','w') as file:
    file.write('Hellow is the the file writing')
    #with 'w' the overwrite the file remove all the exiting text in a file 
```

```python
# writing with append "a" the eiting in new line in a file dont overiting
with open('example.txt','a') as file:
    file.write("\n hwllow  this is my new line")
```

Checking file is exist or not
===


```python
import os 
if os.path.exists('example.txt'):
    print("file found")
else:
    print("file not found")    
    
```

Delecting a file
===

```python
import os
os.remove('del.txt')
```

File position
===

```python
# here is to how to position a file
f=open('example.txt','r')
print(f.tell())  # it will give the current position of the file
f.seek(0)  # it will set the position to the start of the file
```

Reading line by line
===

```python
with open('example.txt','r') as file:
    for line in file:
        print(line)
```
